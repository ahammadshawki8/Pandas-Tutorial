```python
import pandas as pd

df=pd.read_csv("stackoverflow/survey_results_public.csv")
schema_df=pd.read_csv("stackoverflow/survey_results_schema.csv")

pd.set_option("display.max_columns", 85)
pd.set_option("display.max_rows", 85)

df.head()
```

```python
# here we can see that, a dataframe is made of multiple rows and columns.
# if we are only using python, but not using pandas to store information in rows and columns, then how would we do this?
# its a good idea to put all the datas in a dictionary.
```

```python
# this dictionary represents data of a single person
person={
    "first":"Ahammad",
    "last":"Shawki",
    "email":"ahammadshawki8@gmail.com"
}

# this dictionary represents data of multiple people
mul_people={
    "first":["Ahammad","Corey"],
    "last":["Shawki","Schafer"],
    "email":["ahammadshawki8@gmail.com","CoreyMScafer@gmail.com"]
}
# instead of using single value here we are using list containing multiple values.
# we can kindly think them as rows and columns.
# the keys will be columns and the values will be rows.

# if we look at online for the defination of dataframe,
# we can see that many time it is said that dataframe is a two dimentional data structure 
# which basically means rows and columns.
```

```python
# if we want to run that email column we can print that by using the key email
print(mul_people["email"])
```

```python
# dataframe are mostly simmilar to this.
# but it has more functionality that what we have here in standered python.
# we can actually create a dataframe within this dictionary and see what that looks like.
# we can crete a dataframe with a dictionary with DataFrame() function.
dic_df=pd.DataFrame(mul_people)

# if we run this we can see a table like we did before.
dic_df
```

```python
# here the corresponding values of our rows 0, 1 are called index.
# basicaaly indexes are unique values for our rows.

# now we have bit of an idea how to think about dataframes.
# now lets look in how to access informations within the dataframe.

# first, let's access the values of a single column.
# we can access it just like of a dictionary. but instead of using keys here we will use columns name.
print(dic_df["email"])
```

```python
# but a dataframe is much much more than a dictionary or a list.
# for example, we can see that when we printed email column from the dataframe,
# it doesn't look same when we printed the email list using key from the dictionary. 
# and thats because this is actually returning a series 
# and we can see that if we check the type of that email column using type() method.
print(type(dic_df["email"]))
```

```python
# so this is a series object.
# so what is a series?
# A series is basically a list of data.
# but just like as a dataframe it has lot more functionalities than just a list.
# if we look for the series defination in online we can see that-
# in many places, series is defined as a one dimentional array. it is basically means that rows of data of single column.
# we can think of a dataframe as rows and columns and a series as rows of single column.
# so, dataframe is basically a container for multiple of this series objects.
# the series also has a index as well.
```

```python
# as we did earliar, we can print the series just like we print the values using keys.
print(dic_df["email"])
print()
# we can also print the series using dot(.).
print(dic_df.email)

# we could use both method.
# but the first one is better.
# because there is a chance that one of our column name is same to an attribute name of our dataframe.
# if that the case, then using the dot notation might give us some error.
```

```python
# in order to access multiple columns, we can use a list instead of a single string.
print(dic_df[["first","last"]])

# we can also print that without using print.
dic_df[["first","last"]]
# thats the good way of doing that.  
```

```python
# now, that here we are passing here multipe columns, this is no longer a series.
# because a series contains multiple rows in a single column.
# so when we are printing multiple columns it is retuning another dataframe.
# we can see the type of this-
print(type(dic_df[["first","last"]]))
# and in this case, it is a filtered down dataframe with just this specific columns.
# so we have filtered out the "email" column.
# and we can slice this as well simmilar to how we slice a list.
```

```python
# now if we have lots of columns and want to see all of them easily.
# then we can simply grab the columns specificely by saying df.columns
dic_df.columns 
```

```python
# we can see here this gives us all of our columns here.

# now, we have learned how can we grab a column.
# but how can we grab row?
# we can use the loc and iloc indexers
# iloc allows us to access rows by integer locations. iloc means integer locations.
# so if we wanted to get the first row, we can-
dic_df.iloc[0]
# so what that did is that it return a series that contain the values of first row of data.
# whenever we are accessing row like this, our index values will be changed from row to column.
```

```python
# just like we selected multiple columns, we can select multiple rows as well by passing a list of integers
# if we want to access first two rows, we can-
dic_df.iloc[[0,1]]
```

```python
# now, we are getting the dataframe of multiple rows.
# now, with this iloc and loc indexers we can also select columns as well.
# that will be next values of our outer [] brackets.
# if we think of iloc a function, then columns will be our second arguemnet.
dic_df.iloc[[0,1],[1,2]]
# here we cant specify actual column name
# because it is integer location.
# so we can access those values with the integers only.
```

```python
# now, let's look at loc indexer.
# with iloc we were searching by integer locations and with loc we are going to search by labels.
# and we are talking about labels of rows, this will be indexes.
dic_df.loc[[0,1],["email","first"]]
# if we run that, here we are getting email column before first column as we want to print.
# so we can change their order from the dataframe.
```

```python
# now here we see how to grab and access values from our small simple daatset.
# now lets goto our stack over flow  to work with large dataset.

import pandas as pd

df=pd.read_csv("stackoverflow/survey_results_public.csv")
schema_df=pd.read_csv("stackoverflow/survey_results_schema.csv")

pd.set_option("display.max_columns", 85)
pd.set_option("display.max_rows", 85)

df.head()
```

```python
# showing the number of rows and columns of our dataset.
df.shape
```

```python
# seeing what columns are available
df.columns
```

```python
# let's grab the responses from Hobbyiest column
print(df["Hobbyist"])
```

```python
# likely, that doesn't display all of 88883 lines here.
# but we do have a head and tail here to understand what those responses look like.
# let's say we want to know how many of those responses answered yes and how many answers no.
# if we want to do it in python we have counter class or loop to do that.
# but pandas is very powerful that it has so many functions to do that.
# to get the count of unique values in column we can use that value_counts() method to calculate that. 
print(df["Hobbyist"].value_counts())
# we will cover more that in future module when we will analyze data in depth.
```

```python
# now lets grab a specific row and a specific column.
# lets grab the first row and the hobbyiest response for that row using loc indexer.
df.loc[0,"Hobbyist"]
```

```python
# we can also access multiple rows and columns.
# to get the first three responses for hobbyist coloumn we can do that by adding more indexes in list.
df.loc[[0,1,2],"Hobbyist"]
```

```python
# one thing that we havent seen yet is that we can do slicing to get multiple rows and columns.
# if we are fimmiliar with slicing list in python this is going to be same.
# but the only difference is that our last value is going to be inclusive at list with loc.
# inclusuve means our last values will also be added in the slice.
# if we wanted to print the first 3 rows we can -
df.loc[0:2,"Hobbyist"]
# when we are doing slicing, we do not wrap them in brackets[]
```

```python
# we can also do that with column also.
df.loc[[0,1,2],"Hobbyist":"Employment"]
```
